```python
# Task 1: Exploring and Visualizing the Iris Dataset

## 📌 Objective
# To explore and visualize the Iris dataset using pandas, matplotlib, and seaborn.

# ✅ Step 1: Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ✅ Step 2: Load the Iris Dataset
df = sns.load_dataset("iris")

# ✅ Step 3: Dataset Overview
print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nFirst 5 Rows:")
print(df.head())

# ✅ Step 4: Summary Statistics
print("\nSummary:")
print(df.describe())

# ✅ Step 5: Scatter Plot
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()

# ✅ Step 6: Histogram
df["sepal_width"].hist(bins=15)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()

# ✅ Step 7: Box Plot
sns.boxplot(data=df, x="species", y="petal_width")
plt.title("Boxplot of Petal Width by Species")
plt.show()

# ✅ Conclusion
# The visualizations clearly show differences among species based on sepal and petal measurements.
```
